**28_getTrackbarPos() Function in OpenCV**

```python
import cv2
import numpy as np
```

```python
def func(x):
    pass
```

```python
img = np.zeros((500,500,3),np.uint8)  #Create zeros image(Black image)
cv2.namedWindow("bar") #window name
cv2.createTrackbar("on","bar",0,100,func) #Trackbar to which rotate number shown (0 to 100)count, onChange=a function that doesn't do anything by change
while True:
    cv2.imshow("bar", img)
    if cv2.waitKey(1) & 0xff == ord("p"):  #0xff == ord("p")--->  a core thats run if wait key doesnt work
        break
    var = cv2.getTrackbarPos("on","bar")   #get data (output)
    print(var)
    img[:] = [0,0,var]
cv2.destroyAllWindows()
```
