**38_Histograms Equalization in OpenCV**

**Histogram equalization** is a method in image processing of contrast adjustment using the image's histogram.

This method usually increases the global contrast of many images, especially when the usable data of the image is represented by close contrast values.

- Adaptive
- CLAHE (Contrast Limited Adaptive Histogram Equalization)

```python
import cv2
import matplotlib.pyplot as plt
```

```python
### Adaptive

img = cv2.imread(r"C:\Users\hp\Documents\jupyter\OpenCV by WsCube Tech\Images\he.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# hist = cv2.calcHist([img],[0],None,[255],[0,255])
# plt.plot(hist

hist1 = cv2.calcHist([gray],[0],None,[255],[0,255])
plt.plot(hist1)

#Equalizer
eq = cv2.equalizeHist(gray)
hist2 = cv2.calcHist([eq],[0],None,[255],[0,255])
plt.plot(hist2,c="red")

plt.show()

cv2.imshow("Image", img)
cv2.imshow("EQ", eq)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
###  CLAHE


img = cv2.imread(r"C:\Users\hp\Documents\jupyter\OpenCV by WsCube Tech\Images\cl.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist1 = cv2.calcHist([gray],[0],None,[255],[0,255])

#Equalizer
eq = cv2.equalizeHist(gray)
hist2 = cv2.calcHist([eq],[0],None,[255],[0,255])

cl = cv2.createCLAHE(clipLimit=1, tileGridSize=(8,8))
f = cl.apply(gray)

plt.plot(hist1)
plt.plot(hist2, c="red")
plt.plot(f, c="green")

plt.show()

cv2.imshow("Image", img)
cv2.imshow("EQ", eq)
cv2.imshow("CLAHE", f)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
